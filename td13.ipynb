Soit :
$$f(x)=x^2-2x+1=(x-1)^2\quad et\quad g(x)=x+1$$

### Exercice 1 : Méthode analytique
Point d'intersection :
$$
\begin{align*}
f(x) &= g(x)\\
g(x) - f(x) &= 0\\
(x+1)-(x^2-2x+1)&= 0\\
3x-x^2&=0\\
x(3-x)&=0\\
x = 0 \quad &ou \quad x=3
\end{align*}
$$
On cherche $$[a,b]$$ tel que $$g(x) \geq f(x)$$

donc :

$$
a = 0\quad, \quad b = 3
$$
On intègre :
$$
\begin{align*}
A_{exacte} &= \int_0^3 [g(x)-f(x)]dx\\
A_{exacte} &= \int_0^3 (3x-x^2)dx\\
A_{exacte} &= \left[\frac{3x^2}{2}-\frac{x^3}{3}\right]_0^3\\
A_{exacte} &= \frac{3\times3^2}{2}-\frac{3^3}{3}\\
A_{exacte} &= \frac{27}{2}-\frac{27}{3} = \frac{27(3-2)}{6} = \frac{9}{2}
\end{align*}
$$

### Exercice 2 : Tracé sur $[-1,5]$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 5 , 6*100+1) # Domaine du tracé
ab = np.linspace(0, 3 , 6*100+1)
def f(x):
    return x**2 -2*x +1

def g(x):
    return x+1

fig, ax = plt.subplots()
ax.plot(x, f(x))
ax.plot(x, g(x))
ax.fill_between(x, g(x), f(x), where=g(x)>= f(x),color="green")
```
