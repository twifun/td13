# TD 1.3

Soit :
$$f(x)=x^2-2x+1=(x-1)^2\quad et\quad g(x)=x+1$$

### Exercice 1 : Méthode analytique
Point d'intersection :
$$
\begin{align*}
f(x) &= g(x)\\
g(x) - f(x) &= 0\\
(x^2-2x+1)-(x+1)&= 0\\
x^2-3x&=0\\
x(x-3)&=0\\
x = 0 \quad &ou \quad x=3
\end{align*}
$$
On cherche $$[a,b]$$ tel que $$g(x) \geq f(x)$$

donc :

$$
\begin{align*}
g(x) &\geq f(x)\\
g(x) - f(x) &\geq 0\\
(x^2-2x+1)-(x+1)&\geq 0\\
x^2-3x&\geq 0
\end{align*}
$$
