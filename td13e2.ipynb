### Q1

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 3
X = np.linspace(a, b , (b-a)*1000+1) # Domaine du tracé

def f(x):
    return np.exp(-2.5*x**2)

def g(x):
    return (1/(1+x**3/2)-1/10)


fig, ax = plt.subplots()
ax.plot(X, f(X), label=f"$f(x)=e^{{-2.5x^2}}$")
ax.plot(X, g(X), label=f"$g(x)= \\frac{{1}}{{1+\\frac{{x^3}}{{2}}}} - \\frac{{1}}{{10}}$")
ax.fill_between(X, g(X), f(X), where=f(X)>= g(X),color="red")
ax.fill_between(X, g(X), f(X), where=g(X)>= f(X),color="blue")
ax.legend()
```

### Q2

```python
from scipy.optimize import brentq

def trouver_intersection(f, g, a=0, b=3, n=1000):
    X = np.linspace(a, b, (b-a)*n+1)
    phi = f(X) - g(X) 

    indices = np.where(np.diff(np.sign(phi)) != 0)[0]
    return [(X[i], X[i+1]) for i in indices]

print(trouver_intersection(f,g))

#zeros= [brentq(lambda x: f(x)-g(x))]
```
